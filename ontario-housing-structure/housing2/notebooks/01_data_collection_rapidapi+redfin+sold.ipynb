```python

import requests
import httpx
import os
from pathlib import Path 

headers = {
    "x-rapidapi-key": f"{os.getenv('RAPIDAPI_KEY')}",
    "x-rapidapi-host": "redfin-canada.p.rapidapi.com",
    "Content-Type": "application/json"
}
url = "https://redfin-canada.p.rapidapi.com/properties/search-sold"

def get_sold_listings(region_id, sold_within, url=url, headers=headers, max_pages=10):
    DIR_PATH = Path(os.getcwd()).parent / "redfin" / "data_04132026" / f"{region_id}"
    DIR_PATH.mkdir(parents=True, exist_ok=True)

    all_rows = []

    for page in range(1, max_pages + 1):
        querystring = {
            "regionId": region_id,
            "soldWithin": sold_within,
            "limit": 1000,   # API max shown in your response meta
            "page": page
        }

        response = httpx.get(url, headers=headers, params=querystring, timeout=60.0)
        if response.status_code != 200:
            print(f"Error: {response.status_code} - {response.text}")
            return all_rows

        payload = response.json()
        rows = payload.get("data", [])  # <-- key fix

        if not rows:
            break

        all_rows.extend(rows)

        with open(DIR_PATH / f"redfin-listings_results_sold_price_{page}.json", "w", encoding="utf-8") as f:
            json.dump(rows, f, ensure_ascii=False, indent=2)

        with open(DIR_PATH / f"redfin-listings_full_sold_price_{page}.json", "w", encoding="utf-8") as f:
            json.dump(payload, f, ensure_ascii=False, indent=2)

        total_page = payload.get("meta", {}).get("totalPage")
        if total_page and page >= total_page:
            break

    return all_rows
```

```python
DIR_PATH = Path(os.getcwd()).parent / "redfin" / "data_04132026" 
listings_json = sorted(DIR_PATH.rglob("*_results_*.json"))

def load_json_file(path):
    with open(path, encoding="utf-8") as f:
        payload = json.load(f)
    if isinstance(payload, list):
        return pd.json_normalize(payload, sep="_")
    return pd.json_normalize(payload, sep="_")

all_dfs = []
for result in listings_json:
    df_result = load_json_file(result)
    df_result["source_file"] = str(result.relative_to(DIR_PATH))
    all_dfs.append(df_result)

all_df = pd.concat(all_dfs, ignore_index=True, sort=False)
print(f"Loaded {len(listings_json)} files into one DataFrame with shape {all_df.shape}")
all_df.head()
```

```python
# scrub and clean data for analysis
df = all_df.copy()

# 1) dedupe
id_candidates = [c for c in ["homeData_propertyId", "propertyId", "listingId"] if c in df.columns]
if id_candidates:
    df = df.drop_duplicates(subset=id_candidates, keep="first")
else:
    df = df.drop_duplicates()

# 2) numeric coercion
num_cols = [
    "sold_price",
    "homeData_priceInfo_amount",
    "homeData_beds",
    "homeData_bathInfo_computedFullBaths",
    "homeData_sqftInfo_amount",
    "homeData_addressInfo_latitude",
    "homeData_addressInfo_longitude",
]
for c in num_cols:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors="coerce")

# 3) text cleanup
text_cols = [c for c in ["homeData_addressInfo_city", "homeData_addressInfo_state", "homeData_addressInfo_postalCode"] if c in df.columns]
for c in text_cols:
    df[c] = df[c].astype("string").str.strip()

if "homeData_addressInfo_state" in df.columns:
    df["homeData_addressInfo_state"] = df["homeData_addressInfo_state"].str.upper()

# 4) basic validity filters
if "sold_price" in df.columns:
    df = df[df["sold_price"].isna() | (df["sold_price"] > 0)]

lat_col = "homeData_addressInfo_latitude"
lon_col = "homeData_addressInfo_longitude"
if lat_col in df.columns and lon_col in df.columns:
    df = df[
        df[lat_col].isna() | df[lon_col].isna() |
        (df[lat_col].between(-90, 90) & df[lon_col].between(-180, 180))
    ]

# 5) drop very sparse columns (>95% null)
null_frac = df.isna().mean()
drop_sparse = null_frac[null_frac > 0.95].index.tolist()
df = df.drop(columns=drop_sparse)

all_df = df
print(all_df.shape)
```
