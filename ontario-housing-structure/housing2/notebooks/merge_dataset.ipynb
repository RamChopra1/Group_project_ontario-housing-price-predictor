```python
from pathlib import Path 
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_dir = Path.cwd().parent / "data"
datasets = data_dir.rglob("*.csv")
for dataset in datasets:
    
    print(dataset.name)
```

```python
rapid_df = pd.read_csv(data_dir / "raw" / "RapidAPI_Dataset_Sold.csv")
scrape_df = pd.read_csv(data_dir / "raw" / "Scraped_Dataset_Sold.csv")

print(f"RapidAPI Dataset Columns\n{rapid_df.columns}\n")
print(f"Scrape Dataset Columns\n{scrape_df.columns}\n")
```

```python
print("RapidAPI Dataset shape: ", rapid_df.shape)
print("Scrape Dataset shape: ", scrape_df.shape)

print("RapidAPI Dataset INFO: ", rapid_df.info())
print("Scrape Dataset INFO: ", scrape_df.info())
```

```python
rapid_df["mls_number"] = rapid_df["homeData_mlsId"]
rapid_df = rapid_df.drop(columns="homeData_mlsId")

merge_df = pd.merge(rapid_df, scrape_df, on="mls_number", how="left")
merge_df.head()
```

```python
process_dir = data_dir / "processed"
if not process_dir.exists():
    process_dir.mkdir()
merge_df.to_csv(process_dir / "merged_dataset.csv", index=False)
```
